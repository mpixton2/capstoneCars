```python
import pandas as pd
import matplotlib as plt
import seaborn as sns
from sklearn.neighbors import NearestNeighbors
from sklearn.model_selection import cross_val_score
from sklearn.model_selection import train_test_split
from sklearn.model_selection import GridSearchCV
from sklearn.preprocessing import OneHotEncoder
from sklearn.pipeline import Pipeline
import pickle
```

```python
dfcars = pd.read_csv("cars.csv")
dforder = pd.read_csv("orders.csv")
dforder_car = pd.read_csv("orders_cars.csv")
```

```python
df = dfcars.merge(dforder_car, on="car_id")
df = df.merge(dforder, on = "order_id")
print(dforder_car.count())
print(df.count())
df.head()
```

```python
feature_cols = ['car_year', 'car_used','car_mm','car_color']
X = dfcars[feature_cols]
enc = OneHotEncoder(handle_unknown="ignore")
enc.fit(X)
d2 = enc.transform(X)
nn = NearestNeighbors()
nn.fit(d2)
nnresults = nn.kneighbors(d2[0,], n_neighbors = 3)[1]
print((nnresults.tolist()))
# for item in nnresults[1]:
#     print(dfcars.iloc[item,])
```

```python
# Specify the file path to save the pipeline
file_path = 'trained_model.pkl'
file_path2 = 'encoder.pkl'

# Save the pipeline to disk
with open(file_path, 'wb') as file:
    pickle.dump(nn, file)

with open(file_path2, 'wb') as file:
    pickle.dump(enc, file)


```
